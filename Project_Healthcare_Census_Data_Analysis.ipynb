**<u>Problem Statement 3: (Rename State/UT Names)</u>**

**The State/UT names are in all caps in the census data, For uniformity across datasets we use the names so that only the first character of each word in the name is in upper case and the rest are in lower case. However, if the word is and then it should be all lowercase.**

**Examples:** 
* **Andaman and Nicobar Islands**
* **Arunachal Pradesh**
* **Bihar**


```python
# Let's first check all the State/UT names in the census dataset 

census_data_c['State/UT'].value_counts()
```

```python
# Let's change the name format
census_data_c['State/UT'] = census_data_c['State/UT'].apply(lambda x: ' '.join(i.capitalize() for i in x.split(' ')))

# Let's replace 'And' and 'OF' conjunctions with 'and', 'of'
census_data_c['State/UT'] = census_data_c['State/UT'].apply(lambda x: x.replace(' And ',' and ').replace('Nct Of Delhi','Nct of Delhi'))
```

```python
# After transforming the State/UT names Let's check the 'State/UT' column value counts

census_data_c['State/UT'].value_counts()
```

**<u>Problem Statement 4: (New State/UT formation)</u>**
    
* **In 2014 Telangana was formed after it split from Andhra Pradesh, The districts that were included in Telangana are stored in Data/Telangana.txt . Read the text file and Rename the State/UT From Andhra Pradesh to Telangana for the given districts.**

* **In 2019 Laddakh was formed after it split from Jammu and Kashmir, which included the districts Leh and Kargil.  Rename the State/UT From Jammu and Kashmir to Ladakh for the given districts.** 


```python
# Let's read the Telangana.txt file from system
with open('C:/Users/Futurense/Project - Healthcare Census/Telangana.txt ', 'r') as dist:
    telengana_dist = dist.readlines()

# let's remove the \n with space and see the list of telengana district
telengana_dist = [i.replace('\n','') for i in telengana_dist]
print(f"List of telengana districts are :-\n\n{telengana_dist}")
```

```python
# Let's change the districts that were included in Telangana Districts list, let's change those states from Andhra Pradesh to Telangana

census_data_c["State/UT"] = census_data_c.apply(lambda x: 'Telangana' if x['District'] in telengana_dist else x["State/UT"], axis = 1)
```

```python
# Let's check the State/UT value counts if Telangana has been added or not

census_data_c['State/UT'].value_counts()
```

* **Yes! we have successfully added Telangana state.**

```python
# Let's check the rows where we have 'Leh' and 'Kargil' in the district column

census_data_c[census_data_c["District"].apply(lambda x: x.startswith('Leh') or x.startswith('Kargil'))]
```

* **Now, as we can see from above that Leh(Ladakh) is written together, we need only Leh as a district so we will remove 'Ladakh'.**

```python
# Let's change the name of State/UT from 'Jammu and Kashmir' to 'Ladakh' for 'Leh' and 'Kargil' district.

census_data_c.loc[census_data_c["District"].apply(lambda x: x.startswith('Leh') or x.startswith('Kargil')),'State/UT'] = 'Ladakh'
```

```python
# Also we need to change the District name from Leh(Ladakh) to Leh

census_data_c['District'] = census_data_c['District'].replace('Leh(Ladakh)', 'Leh')
```

```python
# Let's check we have successfully changed the state/ut and district names 

census_data_c[census_data_c["District"].apply(lambda x: x.startswith('Leh') or x.startswith('Kargil'))]
```

* **As I had previously already checked the states/UT names in the census dataset and housing dataset, all the states/UT names in both of the dataset are already same, but there are differences in the Hospital dataset and the Census datasets State/UT names so we need to rectify them also.**

```python
# Let's create a function to change the state/ut names in the hospitals dataframe 

def rectify_state_name(df):
    df["State/UT"]=df["State/UT"].str.lower()  
    df["State/UT"]=df["State/UT"].str.title()
    df['State/UT'] = df['State/UT'].str.replace('&','and')
    df['State/UT'] = df['State/UT'].str.replace("*","")
    df["State/UT"]=df["State/UT"].replace({"Odisha":"Orissa","Delhi":"Nct of Delhi","Puducherry":"Pondicherry","India": "All India"})
    return df
```

```python
# Let's assign a variable where updated names will be saved

hospitals = rectify_state_name(hospitals)
```

```python
# Let's fill the null values with 0

hospitals_data = hospitals.fillna(0)

# we are imputing the null values with 0, because it may be the reason that there is no hospital for that division in that particular state.
```

```python
# Let's check the info of the dataframe

hospitals_data.info()
```

* **Community Health Centers, Sub-District/Divisional Hospitals etc these columns are object type we need to rectify it to int.**

```python
# In the dataframe All India states numbers are having commas , we need to replace it by space , so that it can be converted to int

hospitals_data.iloc[-1,1:] = hospitals_data.iloc[-1,1:].apply(lambda x: x.replace(',', ''))
```

```python
# Let's convert the columns object datatype to int datatype

numeric_col_covert = ['Number of Primary Health Centers', 'Community Health Centers','Sub-District/Divisional Hospitals','District Hospitals','Hospitals','HospitalBeds']
hospitals_data[numeric_col_covert] = hospitals_data[numeric_col_covert].astype(int)
```

**<u>Problem Statement 17: (Hospitals required to meet the standards)</u>**
    
**In a hypothetical situation if the government wishes to fill the gap between the available facility and the WHO standards all by itself how many government hospitals may be required in each state or union territory as well as at the national level?**


**Find the average number of beds in a government hospital. Divide it by the gap in the number of beds to reach the standards in the region, and round the number to the nearest integer to find the value.
Represent the findings visually.**


```python
# Let's copy the dataset

# In this case we have to use government hospital dataset, because we have to specifically get the average bed datails of govt hospital
hos_who = government_hospital.copy()

# Let's create a new column name avg number of beds in govt. hospital 
hos_who['Average Number of Beds in Govt. Hospital'] =(hos_who['Rural_Government_Beds']+hos_who['Urban_Government_Beds'])/(hos_who['Rural_Government_Hospitals']+ hos_who['Urban_Government_Hospitals'])

# Let's see the dataframe

hos_who
```
