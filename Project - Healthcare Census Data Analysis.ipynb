**<u>Problem Statement 3: (Rename State/UT Names)</u>**

**The State/UT names are in all caps in the census data, For uniformity across datasets we use the names so that only the first character of each word in the name is in upper case and the rest are in lower case. However, if the word is “and” then it should be all lowercase.**

**Examples:** 
* **Andaman and Nicobar Islands**
* **Arunachal Pradesh**
* **Bihar**


```python
# Let's first check all the State/UT names in the census dataset 

census_data_c['State/UT'].value_counts()
```

```python
# Let's change the name format
census_data_c['State/UT'] = census_data_c['State/UT'].apply(lambda x: ' '.join(i.capitalize() for i in x.split(' ')))

# Let's replace 'And' and 'OF' conjunctions with 'and', 'of'
census_data_c['State/UT'] = census_data_c['State/UT'].apply(lambda x: x.replace(' And ',' and ').replace('Nct Of Delhi','Nct of Delhi'))
```

```python
# After transforming the State/UT names Let's check the 'State/UT' column value counts

census_data_c['State/UT'].value_counts()
```

**<u>Problem Statement 4: (New State/UT formation)</u>**
    
* **In 2014 Telangana was formed after it split from Andhra Pradesh, The districts that were included in Telangana are stored in Data/Telangana.txt . Read the text file and Rename the State/UT From “Andhra Pradesh” to “Telangana” for the given districts.**

* **In 2019 Laddakh was formed after it split from Jammu and Kashmir, which included the districts Leh and Kargil.  Rename the State/UT From “Jammu and Kashmir” to “Ladakh” for the given districts.** 


```python
# Let's read the Telangana.txt file from system
with open('C:/Users/Futurense/Project - Healthcare Census/Telangana.txt ', 'r') as dist:
    telengana_dist = dist.readlines()

# let's remove the \n with space and see the list of telengana district
telengana_dist = [i.replace('\n','') for i in telengana_dist]
print(f"List of telengana districts are :-\n\n{telengana_dist}")
```

```python
# Let's change the districts that were included in Telangana Districts list, let's change those states from Andhra Pradesh to Telangana

census_data_c["State/UT"] = census_data_c.apply(lambda x: 'Telangana' if x['District'] in telengana_dist else x["State/UT"], axis = 1)
```

```python
# Let's check the State/UT value counts if Telangana has been added or not

census_data_c['State/UT'].value_counts()
```

* **Yes! we have successfully added Telangana state.**

```python
# Let's check the rows where we have 'Leh' and 'Kargil' in the district column

census_data_c[census_data_c["District"].apply(lambda x: x.startswith('Leh') or x.startswith('Kargil'))]
```

* **Now, as we can see from above that Leh(Ladakh) is written together, we need only Leh as a district so we will remove 'Ladakh'.**

```python
# Let's change the name of State/UT from 'Jammu and Kashmir' to 'Ladakh' for 'Leh' and 'Kargil' district.

census_data_c.loc[census_data_c["District"].apply(lambda x: x.startswith('Leh') or x.startswith('Kargil')),'State/UT'] = 'Ladakh'
```
